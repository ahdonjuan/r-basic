```python
import numpy as np
import matplotlib.pyplot as plt

def brownian_motion(T, N, mu, sigma, S0):
    dt = float(T) / N
    t = np.linspace(0, T, N+1)
    W = np.random.standard_normal(size=N+1)
    W = np.cumsum(W) * np.sqrt(dt)
    X = (mu - 0.5 * sigma**2) * t + sigma * W
    S = S0 * np.exp(X)
    return t, S

def simular_primas_auto_bm(N, num_simulaciones):
    # Parámetros del modelo BM
    S0 = 1000  # Precio inicial de la prima
    mu = 0.05  # Tasa de crecimiento esperada
    sigma = 0.2  # Volatilidad

    primas = []
    for _ in range(num_simulaciones):
        _, S = brownian_motion(N, N, mu, sigma, S0)  # Simula el precio de la prima con BM
        primas.append(S[-1])
    prima_promedio = np.mean(primas)
    return prima_promedio, primas

# Parámetros
N = 1  # Duración del seguro en años
num_simulaciones = 1000  # Número de simulaciones

# Simulación y cálculo de prima promedio
prima_promedio, primas = simular_primas_auto_bm(N, num_simulaciones)
print("Prima promedio estimada:", prima_promedio)

# Plot del histograma de las primas simuladas
plt.hist(primas, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Histograma de Primas de Seguro de Autos con Brownian Motion')
plt.xlabel('Prima')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```
